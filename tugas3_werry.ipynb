DATA 1

```python
import pandas as pd
data = pd.read_csv("/medals.csv")
data.head()
```

matplotlib

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.show()
```

scatter

```python
data1 = data['Unnamed: 0']
data2 = data['Bronze']
data3 = data['Gold']
data4 = data['Silver']

fig, ax = plt.subplots()
ax.scatter(data1,data2,marker='<',color = 'yellow')
plt.show
```

```python
data1 = data['Unnamed: 0']
data2 = data['Bronze']
data3 = data['Gold']
data4 = data['Silver']

fig, ax = plt.subplots()
ax.scatter(data1,data2,marker='s',color = 'red')
plt.show
```

```python
data1 = data['Unnamed: 0']
data2 = data['Bronze']
data3 = data['Gold']
data4 = data['Silver']

fig, ax = plt.subplots()
ax.scatter(data1,data2,marker='o',color = '#FA8072')
plt.show
```

line plot

```python
data1 = data['Unnamed: 0']
data2 = data['Bronze']
data3 = data['Gold']
data4 = data['Silver']

fig, ax = plt.subplots()
ax.plot(data1,data2,ls='--',color ='green')
plt.show
```

```python
data1 = data['Unnamed: 0']
data2 = data['Bronze']
data3 = data['Gold']
data4 = data['Silver']
 
fig, ax = plt.subplots()
ax.plot(data3,data4,marker='d',ls='--',color ='black')
plt.show
```

```python
data1 = data['Unnamed: 0']
data2 = data['Bronze']
data3 = data['Gold']
data4 = data['Silver']

fig, ax = plt.subplots()
ax.plot(data3,data4,marker='v',color ='blue')
plt.show
```

bar chart

```python
fig, ax = plt.subplots()
ax.barh(data1,data2,color='#8B4513')
plt.show()
```
